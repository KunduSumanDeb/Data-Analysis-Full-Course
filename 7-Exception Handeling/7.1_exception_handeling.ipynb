- FileNotFoundWError: File Not Found
- ValueErroe: Invalid val
- TypeError: Invalid type 

```python
# Error example

a = b
```

```python
# Exception try,exception Block

try:
    a = b
except:
    print("Hey! This variable is not yet assigned.")
```

```python
# Another way to do the same 

try:
    a = b
except NameError as ex:
    print(ex)
```

```python
# Another way to do the same 

try:
    result = 1/0    
except ZeroDivisionError as ex:
    print(ex)
    print("Please enter the denominator greater than zero.")
```

```python
try:
    num = int(input("Enter a number."))
    result = 10/num
except ValueError:
    print("This is not valid number.")
except ZeroDivisionError:
    print("Please enter the denominator greater than zero.")
except Exception as ex:
    print(ex)
    print("Main Exception ot caught here.")
```

```python
# Try, exception, else block

try:
    num = int(input("Enter a number."))
    result = 10/num
except ValueError:
    print("This is not valid number.")
except ZeroDivisionError:
    print("Please enter the denominator greater than zero.")
except Exception as ex:
    print(ex)
else:
    print(f"The result is {result}.")
```

```python
# Try, exception, else block and finally

try:
    num = int(input("Enter a number."))
    result = 10/num
except ValueError:
    print("This is not valid number.")
except ZeroDivisionError:
    print("Please enter the denominator greater than zero.")
except Exception as ex:
    print(ex)
else:
    print(f"The result is {result}.")
finally:
    print("Exception Complete")
```

```python
# File Exception and Exception Handeling

try:
    file = open("example.txt",'r')
    content = file.read()
    print(content)
except FileNotFoundError:
    print("The file dose not exists.")
finally:
    if 'file' in locals() and not file.closed():
        file.close()
        print("file close")
```
