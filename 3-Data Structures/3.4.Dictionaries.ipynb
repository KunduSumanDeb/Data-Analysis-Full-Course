```python
# Example

student = {
    "name": "Suman Deb Kundu",
    "age" : 24,
    "garde": 'A'
}
print(student)
print(type(student))
```

```python
# single key is always used

student = {
    "name": "Suman Deb Kundu",
    "age" : 24,
    "grade": 'A',
    "name": "Payel Pal"
}
print(student)
print(type(student))
```

```python
# Accessing Dictionatry used

student = {
    "name": "Suman Deb Kundu",
    "age" : 24,
    "grade": 'A'
}
print(student)
print(type(student))

print(f"The name of the student : {student['name']}")
print(f"The age of the student : {student['age']}")
print(f"The Grade of the student : {student['grade']}")

```

```python
# Access Using get() method

student = {
    "name": "Suman Deb Kundu",
    "age" : 24,
    "grade": 'A'
}
print(student)
print(type(student))

print(f"The name of the student : {student.get('name')}")
print(f"The age of the student : {student.get('age')}")
print(f"The Grade of the student : {student.get('grade')}")
```

```python
# Modifying Dictionary Elements
# dictionary are mutuable, update or delete elements

# Adding new key
student["age"] = 23
print(student)
student["address"] = "India"
print(student)
```

```python
# Delete key and value pair.

del student["age"]
print(student)
```

```python
# Dictionary Methods

keys = student.keys()
print(keys)

values = student.values()
print(values)

items = student.items()
print(items)
```

```python
# Shallow Copy

student_copy = student
print(student)
print(student_copy)
```

```python
student["name"] = "Payel"
print(student)
print(student_copy)
```

```python
# Hence we perform Shallow copy by using .copy() function 

student_copy1 = student.copy()
print(student)
print(student_copy1)

student["name"] = "Suman"
print(student)
print(student_copy1)
```

```python
# Iterating Over Dictionaries
# We can use loops to iterate over dictionaries, keys, values, or items

# Iterating over keys
for keys in student.keys():
    print(keys)
print()

# Iterate over values
for values in student.values():
    print(values)
print()

#Iterate over key value pairs
for key,values in student.items():
    print(f"{key}: {values}")
```

```python
# Nested Dictionaries

student = {
    "student1": {"name" : "Suman Deb Kundu", "Gender": 'M', "age" :23},
    "student2": {"name" : "Souryadip Mallik", "Gender": 'M',"age": 20},
    "student3": {"name" : "Nirmalya Maity", "Gender": 'M', "age": 21},
    "student4": {"name" : "Kauskik das", "Gender": 'M', "age": 21}
}
print(student)

print(student["student2"]["name"])
print(student["student2"]["age"])
```

```python
# Iterating over nested dictionaries

for student_id,student_info in student.items():
    print(f"{student_id}: {student_info}")
    for key,value in student_info.items():
        print(f"{key}: {value}") 
```

```python
# Dictionary Comprehension

squares = {
    x: x**2 for x in range(11) 
}
print(squares)
```

```python
# Condition dictionary Comprehension

evens = {
    x:x**2 for x in range(11) if x%2 == 0
}
print(evens)
```

```python
# Practical Examples
# Use a dictionary to count the frequency of ewlements in list

numbers = [1,2,3,4,5,6,1,2,3,4,1,2,3,1,2,1,2,3,4,4]
frequency = {}
for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1
print(frequency)
```

```python
# Merge Two dictionaries into one.

dict1 = {"a":1,"b":2}
dict2 = {"b":1,"c":2}
merge_dict = {**dict1, **dict2}
print(merge_dict)
```
