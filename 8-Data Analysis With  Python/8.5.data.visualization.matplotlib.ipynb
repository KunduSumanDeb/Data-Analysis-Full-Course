```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Createe a line plot

plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic Line Plot")
```

```python
# Create a customized line plot

x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x, y, color = "Green", linestyle = '--', marker = 'o',linewidth = 1, markersize = 5  )
plt.grid(True)
```

```python
# multiple Plots

# Sample Data

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize = (9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,'red')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(y1,x,'purple')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,'blue')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(y2,x,'green')
plt.title("Plot 3")
```

#### BAR PLOT

```python
categories = ['A','B','C','D','E']
values = [5,7,3,8,6]

# create a bar plot

plt.bar(categories,values,color = 'purple')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

#### Histograms

```python
# Sample Data 
data = [1,1,2,3,4,5,6,7,7,2,2,2,3,3,4,4,8,8,9,9,9,10]

# Create a histogram 

plt.hist(data,
         bins=range(1, 12),
         color="green",
         edgecolor="orange")
plt.xticks(range(1, 12))
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,3,4,5,6]
plt.scatter(x, y, color = "blue", marker = "x")
```

```python
# Pipe Chart

labels = ['A','B','C','D']
size = [30,20,40,50]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2,0,0,0)

# Create a pie chart
plt.pie(size,explode = explode, labels = labels, colors = colors,autopct = "%1.1f%%")
```

```python
# SALES DATA VISUALIZATION
import pandas as pd
sales_data = pd.read_csv("sales_data.csv")
sales_data.head(5)
```

```python
sales_data.info()
```

```python
# PLOT TOATAL SALES BY PRODUCTS

total_sales_by_product = sales_data.groupby("Product")['Sales'].sum()
print(total_sales_by_product)
```

```python
total_sales_by_product.plot(kind = "bar",color = "teal")
```

```python
# plot sales trend over time 

sales_trend = sales_data.groupby("Date")["Sales"].sum().reset_index()
print(sales_trend)
plt.plot(sales_trend["Date"],sales_trend["Sales"])
```
