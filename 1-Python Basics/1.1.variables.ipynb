```python
# Declaring and assigning Variables

age = 23
height = 5.3
name = "Suman Deb Kundu"
is_student = True

# printing the variables

print("Age is : ",age)
print("Height : ",height)
print("Name : ",name)
print("IS Student : ", is_student)
```

```python
# Naming Convensions
# Variable names should be descriptive
# They must always start with a letter or a '_' and contains letter, number and underscores
# Variables names are case sensitive

first_name = "Suman"
mid_name = "Deb"
last_name = "Kundu"
full_name = first_name + " " + mid_name + " " + last_name
print("Full Name : ", full_name)
```

```python
# Variable types
# Python is dynamically types, type of a variable is determind at runtime

age = 23
height = 5.3
name = "Suman Deb Kundu"
is_student = True
print("Age is type: ",type(age))
print("Height is a type: ",type(height))
print("Name is a type: ",type(name))
print("IS Student is a type: ", type(is_student))
```

```python
# Dynamic Typing
# Python allows the type of a variable to change as the program executes

var = 10 # int
print(var,type(var))

var = "Hello" # string
print(var,type(var))

var = 5.5 # float
print(var,type(var))
```

```python
# input for user

age =  int(input("Enter your age: "))
print(age,type(age))
```

```python
# Simple Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("Sum: ", sum)
print("Difference: ", difference)
print("Product: ", product)
print("Quotient: ", quotient)
```
