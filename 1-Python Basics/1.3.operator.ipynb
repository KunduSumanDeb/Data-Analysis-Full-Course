```python
# Arithmetic Operations

a = 10
b = 5
add = a + b
sub = a - b
mul = a * b
div = b / a
mod = b % a
floor_div = b // a
exp = a ** b 
print("Additon result: ",add)
print("Subtraction result: ",sub)
print("Multiplication result: ",mul)
print("Division result: ",div)
print("Modulus result: ",mod)
print("Floor Division result: ",floor_div)
print("Exponentiation result: ",exp)
```

```python
# Comparison Operators

a = 10 
b = 20
print("a > b: ", a > b)
print("a < b: ", a < b)
print("a >= b: ", a >= b)
print("a <= b: ", a <= b)
print("a == b: ", a == b)
print("a != b: ", a != b)
```

```python
# Logical Operators

x = True
y = False
print("x and y: ", x and y)
print("x or y: ", x or y)
print("not x: ", not x)
```
