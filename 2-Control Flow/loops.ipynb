```python
# while loop

count = 0
while count < 5:
    print(count)
    count += 1
```

```python
count = 0
while count % 2 == 0:
    print(count)
    count += 1
```

```python
# Loop Control Statements
# Break Statement

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# Continue statement

for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# pass statement
# pass opertor is null operator

for i in range(5):
    if i == 3:
        print(f"Pass: {i} value encountered.")
        pass
    print(i)
```

```python
# Nested loops

for i  in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}.")
```

```python
# Example: Calculate the number of N natural numbers using a while and for loop.

# using while loop
n = 10 
sum = 0
count = 1
while count <= n:
    sum = sum + count
    count += 1
print(f"Sum of first {n} natural numbers using while loop is: {sum}.")
```

```python
# Example: Calculate the number of N natural numbers using a while and for loop.

# using for loop
n = 10
sum = 0
for i in range(1, n+1):
    sum = sum + i
print(f"Sum of first {n} natural numbers using for loop is : {sum}.")
```

```python
# Prime Number
n =int(input("Enter the range: "))
for num in range(1, n):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num)
```
