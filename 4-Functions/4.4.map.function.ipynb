```python
# Map 

def square(x):
    return x**2
square(12)
```

```python
number = [1,2,3,4,5,6,7,8,9]
square_number = list[map(lambda x:x**2, number)]
print(f"The numbers : {number} have squares: {square_number}")
print()
number = [1,2,3,4,5,6,7,8,9]
square_number = list(map(lambda x:x**2, number))
print(f"The numbers : {number} have squares: {square_number}")
```

```python
# Multiple Map Iterators
num1 = [1,2,3,4,5,6,7,8]
num2 = [12,23,34,45,56,67,78,89]
multiply = list(map(lambda x,y: x*y, num1, num2))
for i in range(0,8):
    print(f"The product of number: {num1[i]} and {num2[i]} is {multiply[i]}")
```

```python
# Map() to convert a list of strings to integers
# Use map to convert string to iintegers

str_numbers = ['1','2','3','4','5']
int_numbers = list(map(int, str_numbers))
print(int_numbers) # Output: [1,2,3,4,5]
```

```python
words = ['apple','banana','cherry']
upper_words = list(map(str.upper,words))
print(upper_words)
```

```python
def get_name(person):
    return person['name']

people = [
    {'name':'Suman','age':23},
    {'name':'Payel','age':21}
]

list(map(get_name,people))
```
