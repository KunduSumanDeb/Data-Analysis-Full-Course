```python
# Even Odd Function

def even_odd(num):
    if num %2 == 0:
        print("The number is even.")
    else:
        print("The number is odd.")

```

```python
# Calling this function

even_odd(54)
```

```python
# function with multiple parameters

def add(a,b):
    return a+b
result = add(2,4)
print(result)
```

```python
# Default Parameters

def greet(name):
    print(f"Hello {name}")
greet("Suman")
```

```python
# Variable Length Arguments
# Positional and keywords arguments

def print_numbers(*args):
    for number in args:
        print(number)
print_numbers(1,2,3,4,5,6)
print()
print_numbers(1,2,3,4,5,6,"Suman")
```

```python
# Keywords Arguments

def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional argument: {val}")

    for key,value in kwargs.items():
        print(f"Keyword argument- {key}: {value}")
print_details(1,2,3,4,5,"Suman",7,6,8,9,"Payel",name = "Suman Deb Kundu", age = "24", country = "India")
```
