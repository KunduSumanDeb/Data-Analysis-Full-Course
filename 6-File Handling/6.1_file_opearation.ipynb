#### File Operation - Read and Write File

```python
# Read a whole file

with open('example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
# Read a file Line by Line

with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ## Removes the new line character
```

```python
# Writing a file (Overwriting)

with open('example.txt','w') as file:
    file.write('Hello Worls!\n')
    file.write('I am currently learnig Data Analysis!\n')
```

```python
# Writing a file (Without Overwriting)

with open('example.txt','a') as file:
    file.write('My name is Suman Deb Kundu\n')
    file.write('I am student of Heritage Institute Of Technology.\n')
```

```python
with open('example.txt','r') as source_file:
    content = source_file.read()

with open('destination.txt','w') as destination_file:
    destination_file.write(content)
```
